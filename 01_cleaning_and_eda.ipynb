\#\# Section 2\.1 — Data Quality Checks

We now run a series of validation checks before any analysis begins\.
Each check tests a different way the data could be wrong:

1\. Data types — are numbers stored as numbers and text as text?
2\. Categorical values — any typos or unexpected entries?
3\. Missing values — any empty cells?
4\. Duplicate IDs — does each row represent a unique employee?
5\. Age range — does anyone fall outside working age?
6\. Income validity — is anyone's income zero or negative?
7\. Impossible tenure — has anyone worked here longer than their full career?
8\. Constant columns — which fields have no variation?
9\. Performance rating range — are all rating values present?

```
# Confirm Python is reading each column correctly
# Numbers should be int64 or float64; text should be object

df.dtypes

# .dtypes returns the data type of every column
# int64 = whole number (Age, MonthlyIncome)
# object = text/string (Attrition, Department)
# we want this output to match what we'd expect for each column
```

```
# Inspect every text column for typos or unexpected values

cat_cols = ['Attrition', 'BusinessTravel', 'Department', 'EducationField',
            'Gender', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime']

for col in cat_cols:
    print(f"\n{col}:", df[col].unique())

# .unique() returns every distinct value found in that column
# we loop through all 9 categorical columns and print them out
# this is how you catch issues like 'Sales' vs 'sales' vs 'Sales '
# we want clean, consistent values with no duplicates from typos
```

```
# Count nulls in every column

df.isnull().sum()

# df.isnull() returns True wherever a cell is empty
# .sum() counts the True values per column
# every column should return 0
# if anything's > 0, that column needs attention before analysis
```

```
# Each row should represent exactly one employee
# duplicate IDs would inflate headcount and skew attrition rates

dupe_count = df['EmployeeNumber'].duplicated().sum()
print("Duplicated Employee IDs:", dupe_count)

# .duplicated() flags each repeat after the first occurrence
# .sum() counts how many flags there are
# 0 means every Employee ID is unique

# Optional: surface the dupes if any exist
if dupe_count > 0:
    dupes = df[df['EmployeeNumber'].duplicated(keep=False)]
    print(dupes[['EmployeeNumber', 'Age', 'Department']])
else:
    print("No duplicates — clean.")
```

```
# Anyone outside working age (18-65) is a likely data error

age_oor = df[(df['Age'] < 18) | (df['Age'] > 65)].shape[0]
print("Age out of range:", age_oor)

# df['Age'] < 18 — finds everyone under 18
# df['Age'] > 65 — finds everyone over 65
# | means OR — either condition flags the row
# .shape[0] counts how many rows matched
# 0 = no out-of-range ages
```

```
# Nobody should have zero or negative monthly income

bad_income = df[df['MonthlyIncome'] <= 0].shape[0]
print("Negative or zero income:", bad_income)

# <= means "less than or equal to"
# 0 means every employee has a positive income value
```

```
# An employee can't have worked at THIS company longer than their
# entire career — that would be logically impossible

impossible = df[df['YearsAtCompany'] > df['TotalWorkingYears']]
print("Impossible tenure rows:", impossible.shape[0])

# this compares two columns row by row
# Python checks each row: is YearsAtCompany > TotalWorkingYears?
# 0 means every record passes the logic check
```

```
# Columns where every employee has the same value carry no
# predictive information — flag them so we can exclude later

print("StandardHours unique values:", df['StandardHours'].unique())
print("EmployeeCount unique values:", df['EmployeeCount'].unique())
print("Over18 unique values:        ", df['Over18'].unique())

# .unique() returns every distinct value in that column
# StandardHours = [80] for everyone
# EmployeeCount = [1] for everyone
# Over18 = ['Y'] for everyone
# all three will be excluded from any modeling — they have zero variance
```

```
# Confirm what values exist in PerformanceRating
# this matters because if poor performers were rated and exited
# then "attrition" might include performance-managed exits

print("PerformanceRating unique values:", df['PerformanceRating'].unique())

# the scale is documented as 1-4 (1=Low to 4=Outstanding)
# but only values 3 and 4 appear in the data
# this supports the assumption that observed attrition is voluntary
# rather than performance-driven — there are no low ratings to begin with
```

\#\# Section 2\.2 — Target Variable Construction

The \`Attrition\` column is text \(\`'Yes'\` / \`'No'\`\)\. Before any modeling
or aggregation, we convert it to a binary numeric flag where
\`1 = left\` and \`0 = stayed\`\. This single conversion lets us:

\- Calculate attrition rate as a simple \`\.mean\(\)\`
\- Use the column directly as a target variable in modeling
\- Sum the column to get a count of departures

```
# Convert Attrition (text) to Attrition_Flag (binary 1/0)

df['Attrition_Flag'] = (df['Attrition'] == 'Yes').astype(int)

# (df['Attrition'] == 'Yes') returns True / False per row
# .astype(int) converts True → 1 and False → 0
# we keep the original Attrition column too — both are useful

# Confirm the conversion worked
print(df['Attrition_Flag'].value_counts())
print(f"\nOverall attrition rate: {df['Attrition_Flag'].mean()*100:.2f}%")

# value_counts() shows the count of 0s and 1s
# .mean() of a 0/1 column is the proportion of 1s — the attrition rate
# we expect roughly 16% based on the published dataset description
```

The conversion produced \*\*237 leavers\*\* \(1\) and \*\*1,233 stayers\*\* \(0\),
for an overall attrition rate of \*\*16\.12%\*\*\. This 84/16 class imbalance
is a critical methodological consideration that will be addressed
in \*\*Notebook 4 \(Predictive Modeling\)\*\* through SMOTE and threshold
tuning, and through F1\-based scoring rather than accuracy\.

The dataset is now clean, validated, and ready for the descriptive
segmentation analysis in \*\*Notebook 2\*\*\.
